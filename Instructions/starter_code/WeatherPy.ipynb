```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}appid={weather_api_key}&units=imperial&q=" 
number = []
name = []
lat = []
lng = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
temp_max = []

print("City Number: City Name")
for city in cities:
    weather_response = requests.get(query_url + city).json()
    try:
        number.append(weather_response['id'])
        name.append(weather_response['name'])
        print(f"{weather_response['id']}: {weather_response['name']}")
        lat.append(weather_response['coord']['lat'])
        lng.append(weather_response['coord']['lon'])
        temperature.append(weather_response['main']['temp'])
        humidity.append(weather_response['main']['humidity'])
        cloudiness.append(weather_response['clouds']['all'])
        wind_speed.append(weather_response['wind']['speed'])
        temp_max.append(weather_response['main']['temp_max'])
    except KeyError:
        pass
                
    #print(json.dumps(weather_response, indent=4, sort_keys=True))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict = {
    "ID": number,
    "city": name,
    "lat": lat,
    "lng": lng,
    "temp": temperature,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "wind speed": wind_speed,
    "max temp": temp_max
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()

```

```python
weather_data.to_csv("../output_data/cities.csv")
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
north_data = weather_data.loc[weather_data["lat"] > 0, :]
north_data.head()
south_data = weather_data.loc[weather_data['lat'] < 0, :]
south_data.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots

def data_v_lat(x,y):
    plt.scatter(x['lat'], x[y], marker="o", facecolors="green", edgecolors="black", alpha=0.75)
    plt.xlabel('Latitude')
    plt.ylabel(y)
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(x['lat'], x[y])
    regress_values = x['lat'] * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.annotate(line_eq,(x['lat'].min(),x[y].min()),fontsize=15,color="red")
    plt.plot(x['lat'],regress_values,"r-")
    
```

```python
data_v_lat(north_data,'max temp')
plt.title("Maximum Temperature vs. Latitude in the Northern Hemisphere")
#plt.annotate(line_eq,(0,-20),fontsize=15,color="red")
#plt.savefig('../output_data/TempLatN.png')
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
data_v_lat(south_data,'max temp')
plt.title("Maximum Temperature vs. Latitude in the Southern Hemisphere")
plt.savefig('../output_data/TempLatS.png')
plt.show()
```

```python
print("Clearly, in the northern hemisphere maximum temperature for a city decreases as the latitude increases and gets closer to the north pole.")
print("In the southern hemisphere, maximum temperature for a city increases as the latitude increases and gets closer to the equator (lat=0).  It is likely that slope of the regression line in the southern hemisphere is less steep than the one for the northern hemisphere because at this time of the year the northern hemisphere is tilted AWAY from the sun.  Temperature ranges in the northern hemisphere at this time of the year will have a greater change for changes in latitude.")
```
